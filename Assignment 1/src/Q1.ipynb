```python
# Load necessary packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
# Load fb and bg image and convert them to RGB format
fg = cv2.imread("../images/fg.jpg")
fg = cv2.cvtColor(fg,cv2.COLOR_BGR2RGB)
bg = cv2.imread("../images/bg.jpg")
bg = cv2.cvtColor(bg,cv2.COLOR_BGR2RGB)
# Plot the fg, bg images
plt.imshow(fg)
plt.show()
plt.imshow(bg)
plt.show()
```

First, we reduce the fg image's width, height such that it consists only the harry potter image.

```python
# On approximation:
    #Width:[223-330]
    #Height:[8-200]
print(fg.shape)
hp_green = np.array(fg[0:210,220:335])
plt.imshow(hp_green)
```

```python
scale_percent = 300 # percent of original size
width = int(hp_green.shape[1] * scale_percent / 100)
height = int(hp_green.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
resized_hp_green = cv2.resize(hp_green, dim, interpolation = cv2.INTER_AREA)
```

```python
plt.imshow(resized_hp_green)
plt.show()
resized_hp_green.shape
```

```python
l_green = np.array([0, 0,55])
u_green = np.array([60,200,100])
hp_mask = cv2.inRange(resized_hp_green, l_green, u_green)
plt.imshow(hp_mask,cmap='gray')
plt.show()
```

```python
bg_position = bg[20:20+630,200:200+345]
plt.imshow(bg_position)
plt.show()
```

```python
masked_image = np.copy(resized_hp_green)
masked_image[hp_mask != 0] = [0, 0, 0]
plt.imshow(masked_image)
plt.show()

crop_background = np.copy(bg_position)
crop_background[hp_mask == 0] = [0,0,0]
# plt.imshow(crop_background)
# plt.show()

crop_background = crop_background + masked_image
# plt.imshow(crop_background)
# plt.show()

bg[20:20+630,200:200+345] = crop_background
plt.imshow(bg)
plt.show()
```

# Question 1- Part2

Moody also wants to know how many images of size 720 x 576 can be stored
if cloud of size 2 GB is available. Assuming images are stored for displaying on
standard displays, calculate the number of images you can store in the cloud

# Answer:
<center>Image size = 720x576</center>
<center>Total size = 2gb -> 2*1024mb</center>
<center>No. of channels = 3 -> 1 pixel = 3bytes</center>
<center>Size of single image = 720x576x3bytes -> (720x576x3)/(1024*1024)mb</center>
<center>Total number of images = ((2*1024)/(720*576*3))*(1024*1024)</center>
<center>-> <b>ANS:</b>floor(1726.051028807) = <b>1726 images</b></center>

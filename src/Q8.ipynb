```python
# Load necessary packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import pandas
%matplotlib inline
```

```python
locket = cv2.imread("../images/locket.jpeg")
#locket = cv2.cvtColor(locket, cv2.COLOR_BGR2RGB)
locket_r = locket[:,:,0].astype('uint8')
locket_g = locket[:,:,1].astype('uint8')
locket_b = locket[:,:,2].astype('uint8')
```

```python
def gammaTransform(gamma,image):
    maximum = np.max(image)
    minimum = np.min(image)
    imgNorm = (np.array(image)-minimum)/(maximum - minimum)
    trans = imgNorm ** gamma
    out = np.array(trans)*255
    return out
```

```python
out_r = gammaTransform(0.75,locket_r)
out_g = gammaTransform(0.75,locket_g)
out_b = gammaTransform(0.75,locket_b)
plt.imshow(out_r,cmap='gray')
```

```python
plt.imshow(out_g,cmap='gray')
```

```python
plt.imshow(out_g,cmap='gray')
```

```python
out = np.zeros(locket.shape)
out[:,:,2] = out_r.astype('uint8')
out[:,:,1] = out_g.astype('uint8')
out[:,:,0] = out_b.astype('uint8')
out = np.copy(out.astype('uint8'))
plt.imshow(out)
```

```python
vold = cv2.imread("../images/voldemort.jpeg")
vold = cv2.cvtColor(vold, cv2.COLOR_BGR2RGB)
vold_r = vold[:,:,0].astype('uint8')
vold_g = vold[:,:,1].astype('uint8')
vold_b = vold[:,:,2].astype('uint8')
```

```python
out_r = gammaTransform(1.75,vold_r)
out_g = gammaTransform(1.75,vold_g)
out_b = gammaTransform(1.75,vold_b)
```

```python
fig = plt.figure(figsize=[15,15])
axs = fig.add_subplot(131)
axs.imshow(out_r,cmap='gray')
axs.set_title("Red channel gamma transform of image")
axs = fig.add_subplot(132)
axs.imshow(out_g,cmap='gray')
axs.set_title("Green channel gamma transform of image")
axs = fig.add_subplot(133)
axs.imshow(out_b,cmap='gray')
axs.set_title("Blue channel gamma transform of image")
```

```python
out = np.zeros(vold.shape)
out[:,:,0] = out_r.astype('uint8')
out[:,:,1] = out_g.astype('uint8')
out[:,:,2] = out_b.astype('uint8')
out = np.copy(out.astype('uint8'))
```

```python
fig = plt.figure(figsize=[15,15])
axs = fig.add_subplot(121)
axs.imshow(vold)
axs.set_title("Original Image")
axs = fig.add_subplot(122)
axs.imshow(out)
axs.set_title("Gamma Transformed output")
```
